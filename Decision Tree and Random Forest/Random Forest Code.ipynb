we are building a Random Forest Classifier model. our objective is to predict the Quailty of the wine. 

```python
#Importing Libraries for Exploratory Data Analysis
import numpy as np
import pandas as pd

#Importing Libraries for Data Visualization
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
#Importing dataset from a URL.

url = 'http://mlr.cs.umass.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
data = pd.read_csv(url,sep = ';')
```

# Data Preprocessing

```python
data['taste'] = 'Good'
```

```python
data.loc[(data.quality<=5),'taste'] = 'Bad'
data.loc[(data.quality>=6),'taste'] = 'Good'
#data.loc[(data.quality>=5) & (data.quality >=7),'taste'] = 'Normal'
```

```python
sb.countplot(x= 'taste', data=data)
```

```python
data.head(15)
```

```python
# Taking X and Y variables for prediction.
X = data.iloc[:,:10]
y = data.quality
```

```python
# data scaling
from sklearn.preprocessing import scale
X = scale(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.20,random_state = 30)
```

```python
# Machine Learning Model

from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
X_train
```

```python
y_train
```

```python
model.fit(X_train,y_train)
```

```python
predict = model.predict(X_test)
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
 accuracy_score(y_test,predict)
```

```python
confusion_matrix(y_test,predict)
```
